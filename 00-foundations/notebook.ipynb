# Module 00 — Growth curves

Plot the standard growth functions on a log-log axis so the difference between Θ(log n), Θ(n), Θ(n log n), Θ(n²) is immediately visible.

On a log-log plot a function `Θ(n^k)` shows up as a straight line of slope k. Logarithmic growth bends inward.

**Tip:** if anything in this notebook fails, run `pytest 00-foundations/tests -q` first to confirm the module's reference implementations are healthy.

```python
import sys, os
ROOT = os.path.abspath(os.path.join(os.getcwd(), '..'))
if ROOT not in sys.path: sys.path.insert(0, ROOT)

from common.visualize import plot_growth
import math
```

```python
sizes = [10, 100, 1_000, 10_000, 100_000]
series = {
    'log n':       [math.log(n) for n in sizes],
    'sqrt n':      [math.sqrt(n) for n in sizes],
    'n':           [n for n in sizes],
    'n log n':     [n * math.log(n) for n in sizes],
    'n^2':         [n * n for n in sizes],
}
plot_growth(sizes, series, title='Standard growth functions', ylabel='operations (relative)');
```

### Compare two implementations of the same problem

Power: naive `Θ(n)` vs fast `Θ(log n)`. Notice the linear curve climbs much faster on log-log.

```python
from common.bench import sweep
import importlib.util, os

def _load(rel):
    spec = importlib.util.spec_from_file_location('m', os.path.join(ROOT, rel))
    m = importlib.util.module_from_spec(spec)
    spec.loader.exec_module(m)
    return m

fast = _load('00-foundations/problems/03-power-fast/reference.py')

ns = [10, 100, 1_000, 10_000, 100_000, 1_000_000]
linear = [r.best_ms for r in sweep(lambda n: fast.power_linear(2, n), ns)]
fast_  = [r.best_ms for r in sweep(lambda n: fast.power_fast(2, n), ns)]

plot_growth(ns, {'power_linear': linear, 'power_fast': fast_}, title='Power: linear vs fast (best of 3)', ylabel='ms');
```
