# Module 09 — Graph viz

Draw the canonical 10-node graph from `common/testdata/graph_small.json` and show BFS layers from A.

```python
import sys, os, json
ROOT = os.path.abspath(os.path.join(os.getcwd(), '..'))
if ROOT not in sys.path: sys.path.insert(0, ROOT)

from common.visualize import plot_graph
with open(os.path.join(ROOT, 'common/testdata/graph_small.json')) as f:
    data = json.load(f)
adj = {n: [] for n in data['nodes']}
for u, v, _w in data['edges']:
    adj[u].append(v)
plot_graph(adj, title='graph_small.json', directed=True);
```

```python
from importlib import util
spec = util.spec_from_file_location('graph', os.path.join(ROOT, '09-graphs/solutions/graph.py'))
graph = util.module_from_spec(spec); spec.loader.exec_module(graph)
print('BFS layers from A:', graph.bfs('A', adj))
```
