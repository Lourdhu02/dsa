# Module 06 — Sort race

Compare insertion / selection / counting / Python's sorted on the same inputs. Log-log axes.

```python
import sys, os, random
ROOT = os.path.abspath(os.path.join(os.getcwd(), '..'))
if ROOT not in sys.path: sys.path.insert(0, ROOT)

from common.bench import sweep
from common.visualize import plot_growth
from importlib import util
spec = util.spec_from_file_location('sorts', os.path.join(ROOT, '06-sorting-searching/solutions/sorts.py'))
sorts = util.module_from_spec(spec); spec.loader.exec_module(sorts)
```

```python
def make_input(n, seed=0):
    rng = random.Random(seed)
    return [rng.randint(0, 100) for _ in range(n)]

ns = [100, 300, 1000, 3000, 10000]
ins_ms = [r.best_ms for r in sweep(lambda n: sorts.insertion_sort(make_input(n)), ns)]
sel_ms = [r.best_ms for r in sweep(lambda n: sorts.selection_sort(make_input(n)), ns)]
cnt_ms = [r.best_ms for r in sweep(lambda n: sorts.counting_sort(make_input(n)), ns)]
tim_ms = [r.best_ms for r in sweep(lambda n: sorted(make_input(n)), ns)]
plot_growth(ns, {'insertion': ins_ms, 'selection': sel_ms, 'counting': cnt_ms, 'timsort': tim_ms}, title='Sort race');
```
